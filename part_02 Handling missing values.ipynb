# Handling missing values 

Handling missing values

How to handle missing values in machine learning and data science

There are mainly two methods to handle missing values

1.Imputation

2.Dropping

We can not feed machine learning,if dataset contains missing values

```python
#Importing missing values

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Loading the dataset to pandas dataframe

dataset = pd.read_csv('Placement_Data_Full_Class.csv')
```

```python
dataset.head()
```

```python
dataset.shape
```

```python
dataset.isnull().sum()
```

In imputation method we are handling missing values with stastical values those are mean,median,mode

We can these stastical values as Central Tendencies

```python
#Analyse the distribution of data in the salary
sns.displot(dataset.salary, kde=True, bins=20, color='blue')
sns.set_style('whitegrid')  # Optional: Set a plot style
plt.title('Distribution Plot using displot')
plt.show()
```

Most of data points in between  2 lakh to 3lakh and the other side data points are less and have outliers so this curve we 

call it as Skew

we can not use mean for this cases,instead we use median or mode to replace the missing data.

For normal distributions i.e data distributed in equal magnitudes. For such cases we use mean values to replace missing 

data.


```python
# Replacing median values in missing data

dataset['salary'].fillna(dataset['salary'].median(),inplace=True)
dataset.isnull().sum()
```

```python
# Replacing median values in missing data

dataset['salary'].fillna(dataset['salary'].median(),inplace=True)
dataset.isnull().sum()
```

```python
# Replacing mean values in missing data

dataset['salary'].fillna(dataset['salary'].mean(),inplace=True)
dataset.isnull().sum()
```

# Balancing Imbalanced Dataset

Dataset with an unequal class Distribution

Example:

We have diabetes dataset,contains patients of diabetes and non diabetes

If data is imbalanced,more datapoints are diabetes patients and less  data points on non diabetes

i.e 1000 on one class and 100 on other class distribution

S0,we can'nt feed this imbalanced dataset on machine learning model

```python
# importing the Dependencies

import numpy as np
import pandas as pd
```

```python
creditdata = pd.read_csv('creditcard.csv')
creditdata.head()
```

```python
creditdata.tail()
```

```python
# distribution of two classes
creditdata['Class'].value_counts()
```

# This is highly imbalanced dataset

0---> legit transaction

1----> fraudulent transaction

```python
# separating the legit and fraudulent transactions

legit = creditdata[creditdata.Class==0]
fraud = creditdata[creditdata.Class==1]
print(legit.shape)
print(fraud.shape)
```

# Undersampling

Undersampling aims to reduce the number of instances in the majority class to create a more balanced dataset.

Building sample  dataset containing similar distribution of legit and fraudulent transactions.

```python
#No of fraudulent transactions--->492

legit_sample = legit.sample(n=492)
print(legit_sample.shape)
```

```python
#Concatenating two dataframes

new_dataset=pd.concat([legit_sample,fraud],axis=0)
new_dataset.head()
```

```python
new_dataset.tail()
```

```python
new_dataset['Class'].value_counts()
```

# Feature Extraction of Text Data

Feature Extraction of Text data by TF-IDF

Feature Extraction is the mapping from textual data to real valued vector is called Feature Extraction

To convert Text data to numerical data this feature extraction comes here.

Because understanding of numerical data easy for machines.

1.Bag of Words(Bow)

List of unique words in text corpus

2.TF-IDF(Term Frequency-Inverse Document Frequency)

To count the no. of times each word appers in document

# TF-IDF Vectorizer

TF = (Number of times appears in document)/(Number of terms in the document)

IDF = log(N/n)

N is the number of documents

n is the number of documents a term t has appeared in .

The IDF of rare word is high ,where the IDF of a frequent word is low

TF-IDF value of term = TF * IDF

```python
from sklearn.feature_extraction.text import TfidfVectorizer

# Sample text documents
documents = [
    "This is the first document.",
    "This document is the second document.",
    "And this is the third one.",
    "Is this the first document?"
]

# Create a TF-IDF vectorizer
tfidf_vectorizer = TfidfVectorizer()

# Fit and transform the documents into TF-IDF vectors
tfidf_matrix = tfidf_vectorizer.fit_transform(documents)

# Get the TF-IDF feature names (terms)
feature_names = tfidf_vectorizer.get_feature_names_out()

# Convert the TF-IDF matrix to a dense array for easier manipulation
tfidf_matrix_dense = tfidf_matrix.toarray()

# Display the TF-IDF matrix
print("TF-IDF Matrix:")
print(tfidf_matrix_dense)

# Display the feature names
print("\nFeature Names:")
print(feature_names)
```
